```python
import streamlit as st
import tensorflow as tf
import numpy as np

# ·ã®·åà·åπ ·à≠·ãï·àµ ·ä•·äì ·àò·åç·àà·å´
st.set_page_config(page_title="·ã®·àµ·àú·âµ ·âµ·äï·â∞·äì (Sentiment Analysis)", page_icon="üß†")
st.title("üß† ·çä·àç·àù/·àù·à≠·âµ ·åç·àù·åà·àõ ·àò·â∞·äï·â∞·äõ")
st.write("·ä†·àµ·â∞·ã´·ã®·âµ·ãé·äï ·ä®·â≥·âΩ ·ã´·àµ·åà·â°·ç£ ·àû·ã¥·àâ 'Positive' ·ãà·ã≠·àù 'Negative' ·àò·àÜ·äë·äï ·ã≠·äê·åç·à≠·ãé·â≥·àç·ç¢")

# 1. ·àû·ã¥·àâ·äï ·ä•·äì ·â¨·ä≠·â∞·à´·ã≠·ãò·à©·äï ·àò·å´·äï (Cache ·â†·àõ·ãµ·à®·åç ·çç·å•·äê·â±·äï ·ã≠·å®·àù·à´·àç)
@st.cache_resource
def load_assets():
    model = tf.keras.models.load_model("sentiment_bilstm_model.keras")
    vec_model = tf.keras.models.load_model("vectorizer.keras")
    return model, vec_model.layers[0]

try:
    model, vectorizer = load_assets()
    st.success("‚úÖ ·àû·ã¥·àâ ·â†·â∞·à≥·ä´ ·àÅ·äî·â≥ ·â∞·å≠·äó·àç!")
except Exception as e:
    st.error(f"‚ùå ·çã·ã≠·àé·âπ·äï ·àò·å´·äï ·ä†·àç·â∞·âª·àà·àù·ç¶ {e}")

# 2. ·ã®·â∞·å†·âÉ·àö ·åç·â•·ãì·âµ (User Input)
user_text = st.text_area("·ä†·àµ·â∞·ã´·ã®·âµ ·ä•·ãö·àÖ ·ã≠·åª·çâ...", placeholder="·àà·àù·à≥·àå·ç¶ This movie was incredible!")

if st.button("·â∞·äï·âµ·äï"):
    if user_text.strip() != "":
        # ·åΩ·àë·çâ·äï ·ãà·ã∞ ·âÅ·å•·à≠ ·àò·àà·ãà·å•
        vec_text = vectorizer([user_text])

        # ·åç·àù·âµ ·àõ·åç·äò·âµ
        prediction = model.predict(vec_text, verbose=0)[0][0]

        # ·ãç·å§·â±·äï ·àõ·à≥·ã®·âµ
        st.divider()
        if prediction >= 0.5:
            st.balloons() # ·ã∞·àµ·â≥·äï ·àà·àò·åç·àà·åΩ
            st.subheader(f"·ãç·å§·âµ·ç¶ üòä Positive")
            st.progress(float(prediction))
            st.write(f"·ä•·à≠·åç·å†·äù·äê·âµ (Score)·ç¶ {prediction:.2f}")
        else:
            st.subheader(f"·ãç·å§·âµ·ç¶ üòû Negative")
            st.progress(float(prediction))
            st.write(f"·ä•·à≠·åç·å†·äù·äê·âµ (Score)·ç¶ {prediction:.2f}")
    else:
        st.warning("·ä•·â£·ä≠·ãé ·àò·åÄ·àò·à™·ã´ ·åΩ·àë·çç ·ã´·àµ·åà·â°·ç¢")
```
